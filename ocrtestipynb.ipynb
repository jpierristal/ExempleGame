```python
#importer les bibliotheques
import cv2
import easyocr
from matplotlib import pyplot as plt
```

```python
# Charger l'image
image_path = "/content/013.JPG"
image = cv2.imread(image_path)
```

```python
#Initialiseur le detecteur easyocr
reader = easyocr.Reader(['en']) # Langue pour la reconnaissance (par exemple 'en' pour l'anglais)
```

```python
# Lire l'image et extraire le texte
resultats = reader.readtext(image_path)
```

```python
# Afficher le texte extrait
for (bbox, texte, prob) in resultats:
    print(f"Texte : {texte}, Probabilité : {prob:.2f}")
```

```python
# afficher les resultats
for (bbox, texte, prob) in resultats:
    (top_left, top_right, bottom_right, bottom_left) = bbox
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(image, texte, (top_left [0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Boucle for : Cette boucle parcourt chaque élément dans la liste resultats, où chaque élément est un tuple contenant les informations du rectangle englobant (bbox), le texte détecté et la probabilité de la détection.

Destruction du tuple bbox : Le tuple bbox contient les coordonnées des coins du rectangle englobant autour du texte détecté. Les lignes suivantes extraient les coordonnées des coins supérieur gauche (top_left), supérieur droit (top_right), inférieur droit (bottom_right) et inférieur gauche (bottom_left).

Dessin du rectangle : La fonction cv2.rectangle() de la bibliothèque OpenCV dessine un rectangle autour de la zone où le texte a été détecté. Les coordonnées des coins supérieur gauche et inférieur droit ainsi que la couleur et l'épaisseur du rectangle sont spécifiées.

Ajout du texte : La fonction cv2.putText() est utilisée pour ajouter le texte détecté à l'image. Les arguments spécifiés incluent l'image, le texte, la position du coin supérieur gauche du texte, la police, la taille de la police, la couleur et l'épaisseur du texte.

Affichage de l'image : Enfin, la fonction plt.imshow() de Matplotlib est utilisée pour afficher l'image avec les annotations (rectangles et texte ajouté). Le code cv2.cvtColor(image, cv2.COLOR_BGR2RGB) convertit l'image de l'espace de couleur BGR utilisé par OpenCV en espace de couleur RGB utilisé par Matplotlib. Ensuite, plt.show() affiche l'image avec les annotations.

En résumé, ces lignes de code prennnent les résultats de la détection de texte, dessinent des rectangles autour de chaque détection et ajoutent le texte détecté à l'image, puis affichent l'image résultante avec les annotations à l'aide de Matplotlib.
